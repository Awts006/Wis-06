```python
import cv2
import keras
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
import glob
# load Model
from keras.models import load_model
model = load_model('C:/Users/c3000094/jupyternotebook/ake/model_tensorflow/modelTf.h5')
model2 = load_model('C:/Users/c3000094/jupyternotebook/ake/model_tensorflow/model_tensorflow2.h5')
```

```python
def class_number(num):
    if num < 10 : return num
    elif num == 10 : return 0
    elif num == 11 : return 1
    elif num == 12 : return 2
    elif num == 13 : return 3
    elif num == 14 : return 4
    elif num == 15 : return 5
    elif num == 16 : return 6
    elif num == 17 : return 7
    elif num == 18 : return 8
    else  : return 9
```

```python
image_path='C:/Users/c3000094/mile/GrayScale/test2/Sami_full/'
path=image_path+"*.jpg"
files = []
df_Data=pd.DataFrame(columns = ['filename', 'NUMBER','model_pre','model_post'])
for file in glob.glob(path):
    files.append(file)
```

```python
for n in files:
  
    head, tail = os.path.split(n)
    filename = tail.split('.')[0]
    number = tail.split('(')[0]
    print(' ==> ', number)
    
    a= cv2.imread(n)
    #cv_img.append(n)
    s = cv2.cvtColor(a,cv2.COLOR_BGR2GRAY)
    h = cv2.resize(s,(28,28))
    b =h.reshape((1,28,28,1)).astype('float32')/255
    x = model.predict_classes(b)
    y = model2.predict_classes(b)
    newrow= {'filename':filename,'NUMBER':number,'model_pre':class_number(x[0]), 'model_post':class_number(y[0])}
    df_Data = df_Data.append(newrow, ignore_index=True)
```
